## Exploratory Data Analysis

#### Histograms to visualize the distribution of vaccinated and unvaccinated population.



plt. hist - use to create histograms of vaccinated and unvaccinated population.

bin = 20  - parameter indicates how many bins the histogram has.

1e7 - 1 * 10^7 (histogram scaled by a factor of 10 million.

```python
# Calculate summary statistics of vaccinated and unvaccinated population
df[['population_vaccinated', 'population_unvaccinated']].describe()

# Visualize the distribution using histograms
plt.figure(figsize=(10,5))

plt.subplot(2,1,1)
plt.hist(df['population_vaccinated'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Vaccinated Population')
plt.xlabel('Vaccinated Population')
plt.ylabel('Frequency')

plt.subplot(2,1,2)
plt.hist(df['population_unvaccinated'], bins=20, color='red', alpha=0.7)
plt.title('Distribution of Unvaccinated Population')
plt.xlabel('Unvaccinated Population')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

```

The taller the bar, the higher the frequency of data in that range.

Information regarding the data distribution can be gleaned from the shape of histogram. We can see from the distribution of vaccinated population histogram above, it indicates that the distribution's centre is hifted to the right and that there are more lower values. Also, it indicates that the distribution's centre has shifted to the left and that there are more higher value from the distribution of unvaccinated population histogram.

Right-skewed distribution - Distribution of vaccinated population
(The distribution extends further towards higher values because of a tail on the right side, with the majority of observation failing towards the lower end of the range.)

Left-skewed distribution - Distribution of unvaccinated population (The distribution extends further towards lower values because of a tail on the left side, with the majority of observation failing towards the higher end of the range.)

##### Kernal Density Plot

Kernal Density Plot (KDE) gives a smooth estimate of a random variable's probability density function, enabling to observe the features and form of the data distribution.

```python
# Set Seaborn style
sns.set_style("darkgrid")

# Identify numerical colums
num_columns = df.select_dtypes(include=["int64","float64"]).columns

# Plot distribution of each numerical feature
plt.figure(figsize=(14,len(num_columns)* 3))
for idx, feature in enumerate(num_columns,1):
    plt.subplot(len(num_columns), 2, idx)
    sns.histplot(df[feature], kde=True)
    plt.title(f"{feature} | Skewness: {round(df[feature].skew(),2)}")
    
# Adjust layout and show plots
plt.tight_layout()
plt.show()
```

#### Vaccination status group by population, cases and deaths between vaccinated and unvaccinated

```python
df['vaccination_status'] = 'Unknown'
df.loc[(df['vaccinated_cases']> 0) | (df['vaccinated_deaths']> 0), 'vaccination_status'] = 'Vaccinated'
df.loc[(df['unvaccinated_cases']> 0) | (df['unvaccinated_deaths']> 0), 'vaccination_status'] = 'Unvaccinated'

data = df.groupby('vaccination_status').agg({
    'vaccinated_cases':'sum',
    'unvaccinated_cases':'sum',
    'vaccinated_deaths':'sum',
    'unvaccinated_deaths':'sum',
    'population_vaccinated':'sum',
    'population_unvaccinated':'sum'

}).reset_index()
print(data)
```

#### Heatmap

- using seaborn


 - convert 'date' feature to numeric representation and set as index


```python
np.random.seed(42)

data = {
    'date': pd.date_range(start='2022-01-01', periods=10, freq='D'),
    'vaccinated_cases': np.random.rand(10),
    'unvaccinated_cases': np.random.rand(10),
    'vaccinated_deaths': np.random.rand(10),
    'unvaccinated_deaths': np.random.rand(10),
    'population_vaccinated': np.random.rand(10),
    'population_unvaccinated': np.random.rand(10),
    'unvaccinated_hosp': np.random.rand(10),
    'vaccinated_hosp': np.random.rand(10),
}

df = pd.DataFrame(data)

# Convert 'date' to a numeric representation
df['date_numeric'] = df['date'].apply(lambda x: x.toordinal())

# Set the 'date' column as the index
df.set_index('date', inplace=True)


corr_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```
